```python
t1 = [1,2,3]
t2 = [10,20,30]

t1 + t2 # list 끼리의 덧셈
```

```python
t3[2] = 4
t3
```

```python
t1
```

* t1의 값도 바뀌어있다.
* ***t3라는 저장소를 만들어 값만 복사하는 것이 아닌 t3가 t1의 저장소를 바라보게 됨.***

```python
t4 = t1.copy() #copy 옵션으로 복사하여 해결
t4
```

```python
t4[2] = 100
t4
```

```python
t1 # 바뀌어있지 않음
```

# 반복문

```python
for n in [1,2,3]:
    print(n)
```

```python
a = [1,2,3]
for n in a:
    print(a.index(n),"번째 값은 ",n)
```

```python
for n in range(5):
    print(n)
```

```python
for n in range(5):
    if n == 3:
        break
    print(n)
```

```python
for n in range(10):
    if n % 2 == 1 :
        continue
    print(n, "=>", n**2)
```

```python
a = ['aa','bb','cc']
'aa' in a
```

```python
'dd' in a
```

```python
names = ['엘사','안나','스벤','크리스토프','올라프']
peoples = ['엘사','안나','크리스토프']

for each in names:
    if each in peoples:
        print(each, ': 사람입니다.')
    else:
        print(each, ': 사람의 친구입니다.')
```

```python
t = [1,2,3,4,5]
y = []

for n in t:
    y.append(n**2)

y
```

```python
t = [1,2,3,4,5]
y = [n**2 for n in t] # One line commend!!! 가독성이 좋아짐
y
```
